```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd

# Download Recetas de pescado
url = 'https://www.recetasgratis.net/Recetas-de-Pescado-listado_receta-12_1.html'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
response = requests.get(url)
response
```

```python
html = response.content
```

```python
soup1 = bs(html, "lxml")
soup1
```

```python
lista_nombres = []

nombres = soup1.select("div.clear.padding-left-1 div.resultado.link a.titulo.titulo--resultado")

for elem in nombres:
    lista_nombres.append(elem.get_text())

lista_nombres
# len(lista_nombres) 50
```

```python
lista_links = []

links = soup1.select("div.clear.padding-left-1 div.resultado.link a")

for elem in links:
    lista_links.append(elem["href"])

lista_links
# len(lista_links) 50
```

```python
lista_dificultad = []

dificultad = soup1.select("div.clear.padding-left-1 div.resultado.link div.info_snippet span")

for elem in dificultad:
    if "\n" not in elem.get_text():
        lista_dificultad.append(elem.get_text())
    else:
        # lista_dificultad.append(None)
        pass

lista_dificultad
len(lista_dificultad) 
```

```python
lista_personas = []

personas = soup1.select("div.clear.padding-left-1 div.properties span.property.comensales")

for elem in personas:
    if "\n" not in elem.get_text():
        lista_personas.append(elem.get_text())

lista_personas
# len(lista_personas) 48
```

```python
lista_duracion = []

duracion = soup1.select("div.clear.padding-left-1 div.properties span.property.duracion")

for elem in duracion:
    if elem.get_text() == "\n\n\n\n":
        lista_duracion.append(None)
    else:
        lista_duracion.append(elem.get_text())

lista_duracion
len(lista_duracion) 
```

```python
lista_descripcion = []

descripcion = soup1.select("div.clear.padding-left-1 div.intro")

for elem in descripcion:

    lista_descripcion.append(elem.get_text().split("\n")[1])

lista_descripcion
# len(lista_descripcion) 50
```

```python
df_recetas = pd.DataFrame({"nombre":lista_nombres, "dificultad":lista_dificultad, "num_comensales":lista_personas, "duracion":lista_duracion, "descripcion":lista_descripcion, "link":lista_links})
df_recetas
```
